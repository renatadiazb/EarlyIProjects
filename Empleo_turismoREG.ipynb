```python
df_empleoturismo = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/OTHER PROYECTS/turismo_empleo.csv')
df_empleoturismo.head()
```

```python
df_empleoturismo['gasto_promedio_turista'].fillna(df_empleoturismo['gasto_promedio_turista'].mean(), inplace=True)

df_empleoturismo
```

```python
df_empleoturismo.drop(index=19, inplace=True)
```

```python
df_empleoturismo
```

```python
from sklearn.linear_model import LinearRegression
#Definir X y y

X = df_empleoturismo[['gasto_promedio_turista']]
y = df_empleoturismo['empleo_turista']


# Crear modelo de regresin lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(X, y)

# Obtener los coeficientes
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Imprimir los coeficientes
print("Pendiente:", pendiente)
print("Intercepto:", intercepto)

# Hacer predicciones
prediccion = modelo.predict(X)
print("Prediccin:", prediccion)

# Graficar los datos y la lnea de regresin
plt.scatter(X, y, color='blue')
plt.plot(X, prediccion, color='red')
plt.title('Regresin Lineal Simple')
plt.xlabel('Gasto total de turistas de internacin (Dlares)')
plt.ylabel('Empleo Turstico (personas empleadas)')
plt.show()
```

El intercepto (donde mi linea cruza con el eje y) es 3759120, lo cual indica que an sin nada de turismo el empleo inicia en 3,759,120 lo cual es lgico porque el empleo depende de ms variables que de el turismo.

El coeficiente de la pendiente (0.001007611525086658) representa el cambio esperado en la variable de respuesta (en este caso, el empleo turstico) por cada unidad de cambio en la variable predictora (en este caso, el gasto en turismo).

Dado que el coeficiente es positivo, significa que hay una relacin positiva entre el gasto en turismo y el empleo turstico. Especficamente, por cada aumento unitario en el gasto en turismo, se espera un aumento de 0.0.001007611525086658 unidades en el empleo turstico.

Entonces, multiplicando el coeficiente de la pendiente por 1,000,000, obtenemos:

0.001007611525086658 * $1,000 = 1.008

En trminos prcticos, esto podra interpretarse como que a medida que aumenta el gasto en turismo, tambin aumenta el empleo en el sector turstico. Esto puede ser debido a que un mayor gasto en turismo impulsa la demanda de servicios tursticos, lo que a su vez requiere ms mano de obra para cubrir esa demanda.
