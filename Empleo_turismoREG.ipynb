```python
df_empleoturismo = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/OTHER PROYECTS/turismo_empleo.csv')
df_empleoturismo.head()
```

```python
df_empleoturismo['gasto_promedio_turista'].fillna(df_empleoturismo['gasto_promedio_turista'].mean(), inplace=True)

df_empleoturismo
```

```python
df_empleoturismo.drop(index=19, inplace=True)
```

```python
df_empleoturismo
```

```python
from sklearn.linear_model import LinearRegression
#Definir X y y

X = df_empleoturismo[['gasto_promedio_turista']]
y = df_empleoturismo['empleo_turista']


# Crear modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(X, y)

# Obtener los coeficientes
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Imprimir los coeficientes
print("Pendiente:", pendiente)
print("Intercepto:", intercepto)

# Hacer predicciones
prediccion = modelo.predict(X)
print("Predicción:", prediccion)

# Graficar los datos y la línea de regresión
plt.scatter(X, y, color='blue')
plt.plot(X, prediccion, color='red')
plt.title('Regresión Lineal Simple')
plt.xlabel('Gasto total de turistas de internación (Dólares)')
plt.ylabel('Empleo Turístico (personas empleadas)')
plt.show()
```

El intercepto (donde mi linea cruza con el eje y) es 3759120, lo cual indica que aún sin nada de turismo el empleo inicia en 3,759,120 lo cual es lógico porque el empleo depende de más variables que de el turismo.

El coeficiente de la pendiente (0.001007611525086658) representa el cambio esperado en la variable de respuesta (en este caso, el empleo turístico) por cada unidad de cambio en la variable predictora (en este caso, el gasto en turismo).

Dado que el coeficiente es positivo, significa que hay una relación positiva entre el gasto en turismo y el empleo turístico. Específicamente, por cada aumento unitario en el gasto en turismo, se espera un aumento de 0.0.001007611525086658 unidades en el empleo turístico.

Entonces, multiplicando el coeficiente de la pendiente por 1,000,000, obtenemos:

0.001007611525086658 * $1,000 = 1.008

En términos prácticos, esto podría interpretarse como que a medida que aumenta el gasto en turismo, también aumenta el empleo en el sector turístico. Esto puede ser debido a que un mayor gasto en turismo impulsa la demanda de servicios turísticos, lo que a su vez requiere más mano de obra para cubrir esa demanda.

En términos prácticos, esto significa que a medida que el gasto promedio por turista aumenta, también lo hace el nivel de empleo asociado al turismo. Esto puede interpretarse como un indicativo de que un aumento en el gasto turístico podría estar relacionado con una mayor demanda de empleo en el sector turístico.
